```python
import pandas as pd
import numpy as np
import math
from libra.linear_regression import LinearRegressionSMLH
#imports
```

```python
df=pd.read_csv("csvs\HousingData.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df[["CRIM", "ZN", "INDUS", "CHAS", "AGE", "LSTAT"]]=df[["CRIM", "ZN", "INDUS", "CHAS", "AGE", "LSTAT"]].fillna(0)
#selecting the columns with missing values and replacing them with 0
```

```python
df["CRIM"].replace(0, df["CRIM"].mean())
df["ZN"].replace(0, df["ZN"].mean())             
df["INDUS"].replace(0, df["INDUS"].mean())
df["AGE"].replace(0, df["AGE"].mean())
df["CHAS"].replace(0, df["CHAS"].mean())
df["LSTAT"].replace(0, df["LSTAT"].mean())

#replacing the 0 set above with the mean values
```

```python
df.corr()["MEDV"]
```

# Correlation Notes:

Biggest positive correlation with the value of the home is RM - rooms per dewlling, 
because intuitively more rooms equals  bigger surface area, and in turn more expansive and better

Another big but negative correlation is LSTAT - percentage of lower status population, 
because rich people don't want to live near poor people (as they should)

```python
df_corr=df[["INDUS", "RM", "TAX", "PTRATIO", "LSTAT", "MEDV"]]
df_corr
#selecting the columns with the highest correlation
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[df.columns[0:13]], df["MEDV"], test_size=0.3, random_state=56)
```

```python
#training the sklearn model on all data
lr_sklearn=LinearRegression()
lr_sklearn.fit(X_train, y_train)
```

```python
y_pred=lr_sklearn.predict(X_test) #predincting the values, and calculating the errors
print(f"MSE: {mean_squared_error(y_test, y_pred)}")
print(f"MAE: {mean_absolute_error(y_test, y_pred)}")
```

```python
X_train_corr, X_test_corr, y_train_corr, y_test_corr = train_test_split(df_corr[df_corr.columns[1:5]], df["MEDV"], test_size=0.3, random_state=56)
```

```python
#training the sklearn model on selected data
lr_sklearn_corr=LinearRegression()
lr_sklearn_corr.fit(X_train_corr, y_train_corr)
```

```python
y_pred_corr=lr_sklearn_corr.predict(X_test_corr) #predincting the values, and calculating the errors
print(f"MSE correlated: {mean_squared_error(y_test, y_pred_corr)}")
print(f"MAE correlated: {mean_absolute_error(y_test, y_pred_corr)}")
```
