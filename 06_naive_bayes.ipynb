```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("csvs\WA_Fn-UseC_-Telco-Customer-Churn.csv")
data.drop(columns=["customerID"], inplace=True)
data.head(6)
```

```python
data["TotalCharges"]=pd.to_numeric(data["TotalCharges"], errors="coerce")
data.dropna(inplace=True)
```

```python
data.dropna(inplace=True)
data
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder #incoding data
le = LabelEncoder()
for col in data.columns.difference(["SeniorCitizen", "tenure", "MonthlyCharges", "TotalCharges"]):
    data[col] = le.fit_transform(data[col])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=["Churn"]), data["Churn"], test_size=0.3, random_state=765)
#splitting data
```

## Naive Bayes
Training 3 different types of Naive Bayes from sklearn

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB

#creating
GNB=GaussianNB()
MNB=MultinomialNB()
BNB=BernoulliNB()

#training
GNB.fit(X_train, y_train)
MNB.fit(X_train, y_train)  
BNB.fit(X_train, y_train)

#predicting
gnb_pred=GNB.predict(X_test)
mnb_pred=MNB.predict(X_test)
bnb_pred=BNB.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"GNB Acuraccy: {accuracy_score(y_test, gnb_pred)}")
print(f"MNB Acuraccy: {accuracy_score(y_test, mnb_pred)}")
print(f"BNB Acuraccy: {accuracy_score(y_test, bnb_pred)}")
```

```python
scores = pd.DataFrame({"Model": ["GNB", "MNB", "BNB"],
                       "Accuracy": [accuracy_score(y_test, gnb_pred), accuracy_score(y_test, mnb_pred), accuracy_score(y_test, bnb_pred)]})
scores
```

## Conclusion
I conclusion we can say that Gaussian and Bernoulli naive bayes are the best performing, and drawing a parralel to past models trained, naive bayes seems to be much better than logistical regression
